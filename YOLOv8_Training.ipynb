```python
# -----------------------------------------------
# 1. Install YOLOv8
# -----------------------------------------------
!pip install ultralytics

from ultralytics import YOLO
import os

# -----------------------------------------------
# 2. Load YOLOv8 Model
# -----------------------------------------------
model = YOLO("yolov8n.pt")   # nano model for fast training

# -----------------------------------------------
# 3. Train the Model
# -----------------------------------------------
model.train(
    data=r"C:\Users\HP\Desktop\Edureka\Aerial_Object_Classification_Detection\object_detection_dataset\data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    workers=4,
    name="aerial_yolo"
)

# -----------------------------------------------
# 4. Validate Model
# -----------------------------------------------
model.val()

# -----------------------------------------------
# 5. Inference on Test Image
# -----------------------------------------------
results = model("object_detection_dataset/test/images/1.jpg")
results[0].show()
```
